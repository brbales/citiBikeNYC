```python
import pandas as pd
import datetime as dt
```

```python
dtclean_df = pd.read_csv("data/original_data/2017/2017_merged.csv", header=None)
dtclean_df.head()
```

```python
dtclean_df.columns = ["trip_duration", "start_time", "stop_time", "start_station_id", "start_station_name", "start_station_lat", "start_station_lon", "end_station_id", "end_station_name", "end_station_lat", "end_station_lon", "bike_id", "user_type", "birth_year", "gender"]
dtclean_df.head()
```

```python
dtclean_df["start_time"] = pd.to_datetime(dtclean_df.start_time)
dtclean_df["stop_time"] = pd.to_datetime(dtclean_df.stop_time)
dtclean_df["start_time"] = dtclean_df["start_time"].dt.strftime("%m/%d/%Y %H:%M")
dtclean_df["stop_time"] = dtclean_df["stop_time"].dt.strftime("%m/%d/%Y %H:%M")
dtclean_df.head()
```

```python
dtclean_df.to_csv("data/original_data/2017_clean.csv", index=False)
```
